```python
# Importar a biblioteca Pandas e dar um apelido a biblioteca (pd) para facilitar a chamada
import pandas as pd
```

```python
# para armazenar um tabela no python, devemos nomear a tabela e
# alocar nessa variavel o arquivo com a base de dados utilizando a função read

aluguel_casa = pd.read_excel("/content/houses_rent.xlsx")
```

```python
# Para visiualizar a tabela utilizamos a função head
# Por default o pandas mostra as cinco primeiras linha
# Para visulaizar mais linhas, basta colocar a quantidade no head
aluguel_casa.head(10)
```

```python
# Para mostrar a cinco ultimas linhas com o panda usamos a função tail

aluguel_casa.tail(10)
```

```python
# Agora vamos fazer modificações na nossa base de dados
# Para renomear temos a função rename (o que queremos renomear={"o antigo nome":"novo nome"})

aluguel_casa = aluguel_casa.rename(columns={"parking spaces":"parking_spaces",
                                            "rent amount":"rent_amount",
                                            "property tax":"property_tax",
                                            "fire insurance":"fire_insurance"})
```

```python
aluguel_casa.head()
```

```python
# Função dtypes nos mostra os tipos de dados da nossa tabela

aluguel_casa.dtypes
```

```python
# Para alterarmos um tipo de dados da nossa tabela utilizamos outra função no pandas
# a função astype({'o dado que queremos alterar':'o novo tipo que eu quero'})

aluguel_casa = aluguel_casa.astype({"rent_amount":"float",
                                    "property_tax":"float",
                                    "fire_insurance":"float",
                                    "total":"float"})
aluguel_casa.dtypes
```

```python
aluguel_casa.head()
```

```python
# a função info apresenta informações da nossa tabela mais detalhadas
# a coluna Non-Null Count nos ajuda a visualizar se a nossa tabela
# necessita de algum tratamento de dados

aluguel_casa.info()
```

```python
# para visiulizarmos somente as colunas que nós queremos, basta nomearmos as colunas que
# queremos visualizar
# mais ou menos parecido com o que fazemos no SQL, usando o SELECT collumnsName From database
# para descrever as principais métricas das nossas colunas utilizamos a função describe()
# count - contagem
# mean - média
# std - desvio padrao
# min - menor valor
# 25%, 50% e 75% - range que separa os dados entre essas faixas
# max -  maior valor

aluguel_casa[["area", "rooms",
              "bathroom",
              "parking_spaces",
              "rent_amount",
              "property_tax",
              "fire_insurance",
              "total"]].describe()
```

```python
# Filtrando dados com o python
# bastante parecido com o que fazemos no SQL
# Select * From aluguel_casa WHERE total <= 4000
# nome da base de dados[nome da coluna] = filtro

aluguel_casa = aluguel_casa[(aluguel_casa["total"]<=4000)]
aluguel_casa.head()
```

```python
aluguel_casa.describe()
```

```python
# Filtrando por mais parâmetros
# no SQL utilizamos AND
# no Python usamos o &
# para usar a função ou no SQL utilizamos OR
# no Python |

aluguel_casa = aluguel_casa[(aluguel_casa["total"]<=4000)
                            & (aluguel_casa["animal"]=="accept")
                            & ((aluguel_casa["rooms"]==2) | (aluguel_casa["rooms"] == 3))]
aluguel_casa.head()
```

```python
# POdemos analisar agora com o describe a aplicação dos filtros

aluguel_casa.describe()
```

```python
# função query também para fazer filtros

aluguel_casa = aluguel_casa.query("total <= 4000 and animal == 'accept'")
aluguel_casa.head()
```

```python
aluguel_casa = aluguel_casa[(aluguel_casa["total"]<=4000)
                            & (aluguel_casa["animal"]=="accept")
                            & ((aluguel_casa["rooms"]==2)|(aluguel_casa["rooms"]==3))]
aluguel_casa.head()
```

```python
aluguel_casa = aluguel_casa.query(("total <= 4000 and animal == 'accept' and rooms == 2 or rooms == 3"))
aluguel_casa
```

```python
aluguel_casa.describe()
```

```python
# Para agruparmos nossos dados, como faziamos no SQL usamos GROUP BY
# utilizamos a função groupby e a agg
# Agrupamos por cidade utilizando a media do total como valor
# o reset.index() deixa mais vizual nossa análise

agg_city_price = aluguel_casa.groupby(["city"])["total"].agg("mean").reset_index()
agg_city_price
```

```python
# Fazendo mais de um agrupamento
# nome do dados agruapdos = nome dos dados.groupby(as colunas que eu quero agrupar)
# [pelo o que que eu quero agrupar].agg(métrica que eu quero usar)

agg_bath_room_price = aluguel_casa.groupby(["bathroom","rooms"])["total"].agg("mean").reset_index()
agg_bath_room_price
```

```python
# Agora vamos entar na parte visual, plotando gráficos
# para isso utilizamos a biblioteca seaborn

import seaborn as sns
```

```python
# barplot - grafico de barras
# passamos os nossos dados
# depois dizemos qual o eixo x
# depois qual o eixo y
# podemos mudar a cor
# pordemos mudar os rotulos do eixo X e do eixo Y

ax = sns.barplot(data=agg_bath_room_price, x="rooms", y="total", color = "blue");
ax.set(xlabel = "qtd quartos", ylabel = "media do preço");
```

```python
ax = sns.barplot(data=agg_bath_room_price, x="bathroom", y="total", color = "red");
ax.set(xlabel = "qtd banheiros", ylabel = "media do preço");
```

```python
# A coluna floor nos mostra o andar do nosso apartamento
# Quanto mais alto mais valorizado
# Agrupamos os apartamentos pelos andares, para isso utilizamos o nunique para contar os andares
# distintos e o sort para ordenar

aluguel_casa.groupby("floor").index.nunique().sort_values(ascending=False)
```

```python
# Agora vamos agrupar tambem pela média de preço do aluguel

agg_floor_price = aluguel_casa.groupby("floor")["total"].agg("mean").reset_index()
agg_floor_price
```

```python
afloorprice = sns.barplot(data = agg_floor_price, x="floor", y="total", color = "blue");
afloorprice.set(xlabel = "Andar", ylabel = "media do preço");
```

```python
# A partir da minha análise entendi que quanto mais alto o andar mais valorizado,
# logo dos andares mais altos temos os do décimo sexto andar melhor avaliados

sixteenth_floor = aluguel_casa.query("floor == 16")
sixteenth_floor.head()
```
